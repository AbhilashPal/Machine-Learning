```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn 
```

# First let us create our data :

The following are the points around which we shall have our clusters.

```python
X = [30,-100]
Y = [30,-100]
plt.scatter(X,Y)
```

```python
# Let us have 50 points in each cluster
import random

X_train = []
Y_train = []

length = 50

for i in range(length):
    #cluster 1
    x = random.randint(0,50)
    y = random.randint(0,50)
    X_train.append(X[0] + x)
    Y_train.append(Y[0] + y)
    
    #cluster 2
    x = random.randint(0,50)
    y = random.randint(0,50)
    X_train.append(X[1] + x)
    Y_train.append(Y[1] + y)

#plot
plt.scatter(X_train,Y_train)
```

# Kmeans Clustering : 

Firstly we choose two random points in the range (0,100). Thus here, k is 2 and we are clustering the data based on centroid distance(Euclidean distance)

```python
c1x = random.randint(-100,100)
c1y = random.randint(-100,100)
c2x = random.randint(-100,100)
c2y = random.randint(-100,100)
plt.scatter([c1x,c2x],[c1y,c2y])
plt.scatter(X_train,Y_train)
```

```python
def euclidean_distance(x1,y1,x2,y2):
    return np.sqrt(np.square(x1-x2)+np.square(y1-y2))
```

```python
epochs = 5
c1_x = []
c1_y = []
c2_x = []
c2_y = []

length = 50

for _ in range(epochs):
    # for every point we calc which distance is smaller d1 = d(c1,p) or d2 = d(c2,p)
    # The point is put into the class for which d is smaller
    # After this, we calculate the mean of those points and move the centroid there.
    
    for i in range(length*2):
        #point in question
        x = X_train[i]
        y = Y_train[i]
        
        #finding which d is smaller
        if euclidean_distance(x,y,c1x,c1y) < euclidean_distance(x,y,c2x,c2y) :
            c1_x.append(x)
            c1_y.append(y)
        else:
            c2_x.append(x)
            c2_y.append(y)
        
    #we have classified the points. Moving centroid now.
    c1x = np.mean(c1_x)
    c1y = np.mean(c1_y)
    c2x = np.mean(c2_x)
    c2y = np.mean(c2_y)

#plot    
plt.scatter(c1_x,c1_y,c='r')
plt.scatter(c2_x,c2_y,c='g')
plt.scatter(c1x,c1y,c = 'b') 
plt.scatter(c2x,c2y,c = 'b')
```

Thus we see that after 5 epochs, the points are clustered and the centroids move to the respective mean positions of the clusters. We may extend this same concept to n dimensions and create rich labelled datasets from noisy unlabelled data!
